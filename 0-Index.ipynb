# Mini Flash Competition Python Training

# Jake VanderPlas Modified Dr. James Scott

Mini Flash Competition Python Training is a fast-paced introduction to essential components of the Python language for financial students, researchers and developers.

To follow along with this presentation, open a browser and go to [nbviewer](https://nbviewer.org/), enter "jas076100" in the search window. This will take you to this Jupyter Notebook Tutorial "Mini Flash Competition Folder."  Open the folder and select index.ipynb to follow along. After completing this tutorial, you will be able to open the Cookbook_exercise_example.ipynb to perform the operations required for the "Mini Flash Competition."

The material is particularly aimed at those who wish to use Python for financial data analysis and/or scientific programming.
The orginal notebooks are adapted from lectures and workshops by Jake VanderPlas at University of Washington.  Dr. Scott has modified the notebooks for the teaching of financial data analysis.
